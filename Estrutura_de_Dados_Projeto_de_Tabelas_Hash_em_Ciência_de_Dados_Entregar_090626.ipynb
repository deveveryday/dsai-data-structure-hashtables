# Solução Hashtable

```python
products = [[] for _ in range(len(df_produtos))]

def hash_function(value):
  if not isinstance(value, str):
    value = str(value)
  sum_of_chars = 0
  for char in value:
    sum_of_chars += ord(char)
  return sum_of_chars % len(products)

def add(product_details):
  index = hash_function(product_details["nome"])
  products[index].append(product_details)

def contains(product_name):
  index = hash_function(product_name)
  for product_detail_dict in products[index]:
    if product_detail_dict["nome"] == product_name:
      return True
  return False

def get(product_name):
  index = hash_function(product_name)
  found_products = []
  for product_detail_dict in products[index]:
    if product_detail_dict["nome"] == product_name:
      found_products.append(product_detail_dict)
  return found_products if found_products else None


for i in range(len(products)):
    products[i].clear()

for index, row in df_produtos_clean.iterrows():
    product_data = {
        "nome": row["nome"],
        "categoria": row["categoria"],
        "preco": row["preco"]
    }
    add(product_data)




```

# Library

Tentando implementar uma Biblioteca.

Com funções para criar Hash, Listar, Excluir

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None


class HashTableLibrary:
    def __init__(self, capacity):
        if capacity <= 0:
            raise ValueError(">0")
        self.capacity = capacity
        self.size = 0
        self.table = [None] * capacity

    def _hash(self, key):
        if not isinstance(key, str):
            key = str(key)
        sum_of_chars = 0
        for char in key:
            sum_of_chars += ord(char)
        return sum_of_chars % self.capacity

    def add(self, key, value):
        index = self._hash(key)

        if self.table[index] is None:
            self.table[index] = Node(key, value)
            self.size += 1
        else:
            current = self.table[index]
            while current:
                if current.key == key:
                    current.value = value
                    return
                if current.next is None:
                    current.next = Node(key, value)
                    self.size += 1
                    return
                current = current.next


    def get(self, key):
        index = self._hash(key)
        found_values = []

        current = self.table[index]
        while current:
            if current.key == key:
                found_values.append(current.value)
            current = current.next

        if not found_values:
            raise KeyError("")
        return found_values


    def remove(self, key):
        index = self._hash(key)

        previous = None
        current = self.table[index]
        removed_count = 0

        while current:
            if current.key == key:
                if previous:
                    previous.next = current.next
                else:
                    self.table[index] = current.next
                self.size -= 1
                removed_count += 1
                current = self.table[index] if not previous else previous.next
            else:
                previous = current
                current = current.next

        if removed_count == 0:
            raise KeyError("")


    def __len__(self):
        return self.size

    def __contains__(self, key):
        try:
            self.get(key)
            return True
        except KeyError:
            return False

    def display(self):
        for i, bucket in enumerate(self.table):
            if bucket:
                print(f"Bucket {i}:")
                current = bucket
                while current:
                    print(f"{current.key} {current.value}")
                    current = current.next
            else:
                print(f"Empty")

hash_table_capacity = len(df_produtos_clean.drop_duplicates(subset=["nome"])) * 2
product_hash_table = HashTableLibrary(hash_table_capacity)

print(f"capacity: {product_hash_table.capacity}")

for index, row in df_produtos_clean.iterrows():
    product_data = {
        "id_produto": row["id_produto"],
        "nome": row["nome"],
        "categoria": row["categoria"],
        "preco": row["preco"]
    }
    product_hash_table.add(row["nome"], product_data)

example_name_exists = df_produtos_clean.iloc[0]["nome"]
try:
    retrieved_products = product_hash_table.get(example_name_exists)
    print(f"Retrieved {example_name_exists}: {retrieved_products}")
except KeyError as e:
    print(e)


collision_test_name = "Camisa"
try:
    retrieved_products_collision = product_hash_table.get(collision_test_name)
    print(f"Retrieved all {collision_test_name} entries (demonstrating collision handling): {retrieved_products_collision}")
except KeyError as e:
    print(e)


name_to_remove = "Camisa"
try:
    product_hash_table.remove(name_to_remove)
    print(f"Current size after removal: {len(product_hash_table)}")
    try:
        print(f"Attempt to get {name_to_remove} after removal: {product_hash_table.get(name_to_remove)}")
    except KeyError as e:
        print(f"As expected {name_to_remove} not found after removal: {e}")
except KeyError as e:
    print(e)

```
